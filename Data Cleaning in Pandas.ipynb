```python
import pandas as pd
```

```python
# reading an excel file using pd.read_excel
df = pd.read_excel("C:\\Users\\91740\\Desktop\\Python\\05_Data_Analysis_with_Python_for_Excel_users\\Practise data\\Customer Call List.xlsx")
df
```

### Removing duplicates

```python
df = df.drop_duplicates()
df
```

### Remove columns

```python
# to remove multiple columns, column names will be passed as a list
df = df.drop(columns = "Not_Useful_Column")
df
```

### Strip Last_Name column

```python
# strip Last_Name column to remove '_', '.', and '/' from the names
```

```python
df['Last_Name'] = df['Last_Name'].str.lstrip('/')
df['Last_Name'] = df['Last_Name'].str.lstrip('.')
df['Last_Name'] = df['Last_Name'].str.rstrip('_')
# removed unwanted characters from the names and assigned it back to df['Last_Name'] column
df
```

```python
# 2nd syntax for the same operation
df['Last_Name'] = df['Last_Name'].str.strip('/.123_')
df
```

### Cleaning/Standardize Phone_Number

```python
# using replace method()
# here using the pandas replace method, not the string replace method
# string replace method has a simpler syntax
# help(str.replace)
# help(df.replace)
# both has different syntax
df['Phone_Number']
```

```python
# replace all characters other than [^a-zA-Z0-9]
# df['Phone_Number'].str.replace('[^a-zA-Z0-9]','') -- using str replace method
# use reges = True when using reg expresiion
df['Phone_Number'].replace(to_replace = '[^a-zA-Z0-9]', value = '', regex = True, inplace = True)
df
```

```python
# apply method is applied one row at a time

# befor apply method convert the df['Phone_Number'] values to string
df['Phone_Number'] = df['Phone_Number'].apply(lambda x : str(x))

# apply method and split the phone number in the right format
df['Phone_Number'] = df['Phone_Number'].apply(lambda x : x[0:3] + '-' + x[3:6] + '-' + x[6:])
df['Phone_Number']
```

```python
# remove Na-- and nan--
df['Phone_Number'] = df['Phone_Number'].replace(to_replace = 'Na--', value = '')
df['Phone_Number'] = df['Phone_Number'].replace(to_replace = 'nan--', value = '')
df['Phone_Number']
```

```python
# df with transformed 'Phone_Number' column
df
```

### Splitting a column

```python
# split address column and create three seperate columns
# help(str.split)

df[['Street_Address','State','Zip_Code']] = df['Address'].str.split(',',2, expand = True)
df[['Street_Address','State','Zip_Code']]
```

```python
# df with 3 news columns from the address column
df
```

### Standardize column values using replace

```python
# columns 'Paying Customer' and 'Do_Not_Contact'
# N Y replaced with No and Yes
# N/a removed
# NaN removed
```

```python
# replace N with No
df[['Paying Customer','Do_Not_Contact']] = df[['Paying Customer','Do_Not_Contact']].replace(to_replace = 'N', value = 'No')

# replace Y with Yes
df[['Paying Customer','Do_Not_Contact']] = df[['Paying Customer','Do_Not_Contact']].replace(to_replace = 'Y', value = 'Yes')

# replace N/a with ''
df[['Paying Customer','Do_Not_Contact']] = df[['Paying Customer','Do_Not_Contact']].replace(to_replace = 'N/a', value = '')

# replace NaN with ''
# the below replace will not work
# because basically its blank (Not a Number)
#df[['Paying Customer','Do_Not_Contact']] = df[['Paying Customer','Do_Not_Contact']].replace(to_replace = 'NaN', value = '')

# so instead we have to fill that na using fillna method
df = df.fillna('')

df[['Paying Customer','Do_Not_Contact']]
```

```python
# df after Yes and No transformation
df
```

### Fill na / NaN with blank

```python
df = df.fillna('')
df
```

### Filtering down rows of data

#### 1) Method 1

```python
# get the details of customers whom we can contact
# ie. Do_Not_Contact != Y
# method 1
df.loc[df['Do_Not_Contact']!='Yes']
```

#### 2) Method 2

```python
# using for loop and if
```

```python
# remove all records where 'Do_Not_Contact' == 'Yes' and blank
for i in df.index:
    if df.loc[i,'Do_Not_Contact'] == 'Yes' or df.loc[i,'Do_Not_Contact'] == '' :
        df.drop(i, inplace = True)
df    
```

```python
# remove all records where there are no phone numbers
for i in df.index:
    if df.loc[i,'Phone_Number'] == "":
        df.drop(i, inplace = True)
df
```

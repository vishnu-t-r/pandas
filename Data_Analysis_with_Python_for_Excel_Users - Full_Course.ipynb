## 11.2 pivot_table() method

# 12 Data Visualization with Pandas

```python
# importing new data for visualization
import pandas as pd
df_population_raw = pd.read_csv("C:\\Users\\91740\\Desktop\\Python\\05_Data_Analysis_with_Python_for_Excel_users\\Practise data\\population_total.csv")
```

```python
df_population_raw
```

```python
# remove null values from the table
# using dropna() method
df_population_raw.dropna(inplace = True)
```

```python
# using pivot() method
# pivot population table, such that year is index, country as seperate columns and poputaion as aggreagate(value)
df_population_raw.pivot(index = 'year',
                       columns = 'country',
                       values = 'population')
```

```python
# using pivot_table() method
# pivot population table, such that year is index, country as seperate columns and poputaion as aggreagate(value)
df_pop_pivot = df_population_raw.pivot_table(index = 'year',
                              aggfunc = 'sum',
                       columns = 'country',
                       values = 'population')
```

```python
df_pop_pivot.head(2)
```

```python
# selecting only some countries 'United States','India','China','Indonesia','Brazil'
df_pivot = df_pop_pivot[['United States','India','China','Indonesia','Brazil']]
```

```python
df_pivot.head(5)
```

```python
# pandas.DataFrame.plot() - make plot of series or DataFrame
```

## 12.1 line chart

```python
df_pivot.plot(kind = 'line', 
              xlabel = 'Year (1955-2020)', 
              ylabel = 'Population in 100M', 
              title = 'Population (1955-2020)',
              figsize = (7,3))
```
